```python
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("https://cocl.us/datascience_survey_data",index_col=0)
```

```python
df1
```

```python
df2 = df1.sort_values(by = 'Very interested', ascending = False) 
```

```python
df2
```

```python
df2 = (100. * df2/2233).round(2)
df2.head(6)
```

```python
ax=df2.plot(kind='bar',figsize=(20,8),width=0.8,color=['#5cb85c','#5bc0de','#d9534f']) 
plt.legend(fontsize=14)
plt.title("Percentages of Respondents' Interest in Datascience areas",fontsize=16)
plt.xticks(fontsize=14)
plt.yticks([])
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])
for p in ax.patches:
  width, height = p.get_width(), p.get_height()
  x, y = p.get_xy() 
  ax.annotate('{}%'.format(height), (x, y + height + 1))
    
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```
